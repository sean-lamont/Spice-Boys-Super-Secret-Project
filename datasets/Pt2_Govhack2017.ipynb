```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('newDataOut.csv')
```

```python
df
```

```python
df['Recreation&Fitness'] = df['Skate_Parks'] + df['Playgrounds']+df['Basketball_Courts']
df['Recreation&Fitness'] += 10
denom = max(df['Recreation&Fitness'])
df['Recreation&Fitness'] = df['Recreation&Fitness'] / denom
df.to_csv('Out2.csv')
df
```

```python
df['crimesPP'] = (df['crimes'] + 10) / df['population']
max(df['crimes'])
df.to_csv('Out2.csv')
df.set_index('Suburb', inplace= True)
```

```python
df['Bus_Stops'] = 0
BusStops = pd.read_csv('Bus_Stops.csv')
BusStops["Suburb"] = BusStops["Suburb"].map(str.strip)
for index,row in BusStops.iterrows():
    if row['Suburb'] == "MacGregor" :
        df.loc['Macgregor'] += 1
    elif row['Suburb'] == "O'malley" :
        df.loc["O'Malley"] += 1
    elif row['Suburb'] == "O'connor" :
        df.loc["O'Connor"] += 1
    else:
        if row['Suburb'] not in [x.upper() for x in df.index]:
            print (row['Suburb'])
        else:
            matching_index = [x for x in df.index if x.upper() == row['Suburb']][0]
            #print (matching_index)
            df.loc[matching_index, 'Bus_Stops'] += 1
```

```python
df
```

```python
df['Fitness_Sites'] = 0
Fitness = pd.read_csv('Fitness_Sites.csv')
Fitness["DIVISION"] = Fitness["DIVISION"].map(str.strip)
for index,row in Fitness.iterrows():
    if row['DIVISION'] == "MacGregor" :
        df.loc['Macgregor'] += 1
    elif row['DIVISION'] == "O'malley" :
        df.loc["O'Malley"] += 1
    elif row['DIVISION'] == "O'connor" :
        df.loc["O'Connor"] += 1
    else:
        if row['DIVISION'] not in df.index:
            print (row['DIVISION'])
        else:
            df.loc[row['DIVISION'], 'Fitness_Sites'] += 1
```

```python
df
```

```python
df['Unemployment'] = 0
unemploy = pd.read_csv('unemployment.csv')
unemploy = unemploy[['Suburb','Unemployment']]
for index,row in unemploy.iterrows():
    if unemploy.loc[index,'Suburb'] in df.index:
        df.loc[row['Suburb'], 'Unemployment'] = row['Unemployment']
for index,row in df.iterrows():
    #print (row['Unemployment'])
    if np.isnan(row['Unemployment']):
        print (row['Suburb.1'])
        df.loc[index,'Unemployment'] = 3.1
    if row['Unemployment'] == 0:
        print (row['Suburb.1'])
        df.loc[index,'Unemployment'] = 4.1
```

```python
df.to_csv('Out2.csv')
```

```python
df['Transport'] = df['Bus_Stops'] / sum(df['Bus_Stops']) + df['bus_to_TC'] / sum(df['bus_to_TC']) + df['time_to_Hospital'] / sum(df['time_to_Hospital'])
```

```python
df['Education'] = (df['schools'] + 3 )/ sum(df['schools'])*1000 + df['bus_to_TC'] / sum(df['bus_to_TC']) * 100
```

```python
df
```

```python
import math

def sigmoid(x):
  return 1.0 / (1.0 + math.exp(-x))
sigmoid = np.vectorize(sigmoid)
```

```python
df['Livability'] = -100*df['crimesPP'] + 10*df['Recreation&Fitness'] - df['population']/1000.0 + \
    df['Education'] / 10.0 - 5 * df['Unemployment'] + 100*df['Transport']
df['Livability'] = ((df['Livability']/100).apply(sigmoid) + 1)
df['Livability'] = 10000 * df['Livability'] / sum (df['Livability'])
```

```python
df.to_csv('Out2.csv')
df
```
